```python
import numpy as np
```

```python
inp = [[6,7,8,9,1,0], [2,4,5,6,7,0]]
inp = [[[i] * 7 for i in each] for each in inp]
inp = np.array(inp, dtype=np.float32)
sequence_shape = inp.shape
batch_size = sequence_shape[0]
seq_length = sequence_shape[1]
width = sequence_shape[2]
inp = np.expand_dims(inp, 1)
inp = np.tile(inp, [1, seq_length, 1, 1])
inp = np.reshape(inp, [-1, seq_length, 7])
```

```python
positions = np.array([list(range(seq_length)) for _ in range(batch_size)], dtype=np.int32)  # (b, l_t)
positions = np.reshape(positions, -1)
positions = np.expand_dims(positions, 1)
positions
for i in range(batch_size * seq_length):
    inp[i][positions[i]] = [0.5] * 7
positions = np.array([list(range(seq_length)) for _ in range(batch_size)], dtype=np.int32)  # (b, l_t)
positions = np.reshape(positions, -1)
```

```python
flat_offsets = seq_length + 1
```

```python
flat_positions = positions * flat_offsets
```

```python
flat_sequence_tensor = np.reshape(inp, [batch_size * seq_length * seq_length, width])
flat_sequence_tensor
```
